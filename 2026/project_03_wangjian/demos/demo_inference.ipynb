# YOLOv11 + SAM 2 Automatic Annotation Demo

This notebook demonstrates the core inference pipeline for the final project: YOLOv11 detects desktop objects, SAM 2 converts each bounding box into a mask, and the script exports visualized annotations plus COCO-style JSON.

## 1. Check Environment

```python
import sys
from pathlib import Path

cwd = Path.cwd().resolve()
PROJECT_ROOT = cwd.parent if cwd.name == 'demos' else cwd
if not (PROJECT_ROOT / 'src' / 'main.py').exists():
    raise RuntimeError(f'Cannot locate project root from {cwd}')
print(PROJECT_ROOT)
sys.path.append(str(PROJECT_ROOT / 'src'))

try:
    import torch
    print('CUDA available:', torch.cuda.is_available())
except Exception as exc:
    print('Torch is not installed yet:', exc)
```

## 2. Run Sample Inference

The command below processes the test images in `data/test_examples`. If YOLOv11 or the SAM 2 checkpoint is not installed yet, the code will use fallback logic only for smoke testing. Final reported results should be generated with real YOLOv11 and SAM 2 weights.

```python
import subprocess

cmd = [
    sys.executable,
    str(PROJECT_ROOT / 'src' / 'main.py'),
    '--input', str(PROJECT_ROOT / 'data' / 'test_examples'),
    '--output', str(PROJECT_ROOT / 'results')
]
print(' '.join(cmd))
subprocess.run(cmd, check=True)
```

## 3. Show Annotated Outputs

```python
from IPython.display import display
from PIL import Image

figure_dir = PROJECT_ROOT / 'results' / 'figures'
for path in sorted(figure_dir.glob('*_annotated.jpg'))[:3]:
    print(path.name)
    display(Image.open(path))
```
