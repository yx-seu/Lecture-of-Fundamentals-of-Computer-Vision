# Inference Demo

This notebook demonstrates the core inference workflow for the SegFormer Oxford-IIIT Pet segmentation project.

## Steps

1. Prepare a trained model directory
2. Put 2-3 images into `data/test_examples/`
3. Run `src.main` for batch inference
4. Inspect predicted masks and overlay visualizations

```python
import os
from pathlib import Path

project_root = Path.cwd().resolve().parent if Path.cwd().name == 'demos' else Path.cwd().resolve()
model_dir = project_root / 'outputs' / 'segformer_b0_pet' / 'best_model'
input_dir = project_root / 'data' / 'test_examples'
output_dir = project_root / 'results' / 'demo_inference'

print('model_dir =', model_dir)
print('input_dir =', input_dir)
print('output_dir =', output_dir)
```

```python
import subprocess
import sys

cmd = [
    sys.executable,
    '-m', 'src.main',
    '--model_dir', str(model_dir),
    '--input', str(input_dir),
    '--output', str(output_dir),
]
print('Running:', ' '.join(cmd))
subprocess.run(cmd, check=False)
```

```python
from IPython.display import display
from PIL import Image

if output_dir.exists():
    for subdir in sorted(output_dir.iterdir()):
        if subdir.is_dir():
            print('\nExample:', subdir.name)
            for name in ['pred_mask.png', 'overlay.png']:
                path = subdir / name
                if path.exists():
                    display(Image.open(path))
```
